```python
from langchain_chroma import Chroma
from langchain_community.embeddings import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

vector_store = Chroma(
    collection_name="my_collection",
    embedding_function=embeddings,
    persist_directory="./chroma_langchain_db"  # optional
)

# 3 Add your text chunks (texts must be a list of strings)
vector_store.add_texts(texts)

print(vector_store._collection.get())
```

```python
retriever=vector_store.as_retriever()
```

```python
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_messages([
    ("system",
     "You are a precise and factual assistant. "
     "Use the provided context if it contains the answer. "
     "If the context lacks sufficient information, use your own knowledge to give a complete and accurate answer. "
     "Do not mention the context, or say that its missing information. "
     "Always provide a direct, self-contained answer."),
    ("user",
     "Question: {topic}\n\nContext:\n{context}")
])
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from google.colab import userdata

llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    api_key=userdata.get('GEMINI_API_KEY'),
    temperature=0.0
)
```

```python
def formatDocs(docs):
  return "\n\n".join([doc.page_content for doc in docs])
```

```python
from langchain_core.runnables import RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser
rag_chain=({"context":retriever |formatDocs,"topic":RunnablePassthrough()} | prompt | llm | StrOutputParser())
```
